## What is Monte Carlo Simulation
### A Monte Carlo simulation is a type of computational algorithm that estimates the probability of occurrence of an undeterminable event due to the involvement of random variables. The algorithm relies on repeated random sampling in an attempt to determine the probability. This means simulating an event with random inputs a large number of times to obtain your estimation.

## Learning Case 

### Our simple game will involve two six-sided dice. 
### In order to win, the player needs to roll the same number on both dice.
### A six-sided die has six possible outcomes (1, 2, 3, 4, 5, and 6). With two dice, there is now 36 possible outcomes (1 and 1, 1 and 2, 1 and 3, etc., or 6 x 6 = 36 possibilities). In this game, the house has more opportunities to win (30 outcomes vs. the player’s 6 outcomes), meaning the house has the quite the advantage. Let’s say our player starts with a balance of 1,000 and is prepared to lose it all, so they bet 1 on every roll (meaning both dice are rolled) and decide to play 1,000 rolls. Because the house is so generous, they offer to payout 4 times the player’s bet when the player wins. For example, if the player wins the first roll, their balance increases by 4, and they end the round with a balance of 1,004. If they miraculously went on a 1,000 roll win-streak, they could go home with 5,000. If they lost every round, they could go home with nothing.

```python
#importing libraries
import matplotlib.pyplot as plt
import random
```

```python
#Dice Roll Function
def rolldice():
    die1 = random.randint(1,6)    #generate a random integer between 1 and 6
    die2 = random.randint(1,6)
    if die1 == die2:
        same = True
    else:
        same = False
    return same
```

##### Every Monte Carlo simulation will require you to know what your inputs are and what information you are looking to obtain.

```python
#Inputs
simulation = 100000
max_rolls = 500
bet = 1
#Tracking
win_prob = []
end_bal = []
```

```python
#Creating simulation figure

fig = plt.figure()
plt.title('Monte Carlo Simulation for Dice game')
plt.xlabel('Roll')
plt.ylabel('Balance [$]')
plt.xlim([0,max_rolls])
```

# Simulation

#### Process
Outer loop to iterate through pre-defined number of simulations,
inner loop to run each game of 1000 rolls,
A loop for balance and count for wins and rolls.

```python
for i in range(simulation):
    balance = [1000]
    num_rolls = [0]
    wins = 0
    
    while num_rolls[-1] < max_rolls:
        check_same = rolldice()
        if check_same:
            balance.append(balance[-1] + 4*bet)
            wins += 1
        else:
            balance.append(balance[-1] - bet)
            
        num_rolls.append(num_rolls[-1]+1)
        
#storing variables and plotting

    win_prob.append(wins/num_rolls[-1])
    end_bal.append(balance[-1])
    plt.plot(num_rolls, balance)
```

```python
plt.show()

#Avearging win probability and end balance
overall_win_probability = sum(win_prob)/len(win_prob)
overall_end_balance = sum(end_bal)/len(end_bal)

# Displaying the averages
print("Average win probability after " + str(simulation) + " runs: " + str(overall_win_probability))
print("Average ending balance after " + str(simulation) + " runs: $" + str(overall_end_balance))
```
